```python
import cv2
import os
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt

# Directory paths for generated fake images and real reference images
generated_images_dir = r'C:\Users\Manohar Vemuri\OneDrive\MS\Computer Vision\cs512-s23-vemuri-saimanohar\project\data\results\U-Net_CutMix_GAN/'
# real_images_dir = "D:\MS\sem2\cv\proj\00115.png"
real_image=cv2.imread(r"C:\images1024x1024\00001.jpg")
real_image = cv2.resize(real_image, (16, 16))
real_image = cv2.cvtColor(real_image, cv2.COLOR_BGR2GRAY)

# Iterate through all files in the generated images directory
a=[]
d=dict()

for file in os.listdir(generated_images_dir):
    if file.endswith('.png') or file.endswith('.jpg'):
        # Load the generated fake image
        fake_image = cv2.imread(os.path.join(generated_images_dir, file), cv2.IMREAD_GRAYSCALE)

        # Load the corresponding real reference image
        #print(real_image.shape)
        

        # Resize the images to a consistent size
        fake_image = cv2.resize(fake_image, (16, 16))
      

        # Calculate SSIM
        ssim_score = ssim(fake_image, real_image)

        # Print the SSIM score for each pair of images
        a.append(ssim_score)
        d[file]=ssim_score
            
            


max_score=max(d.values())
min_score=min(d.values())
```

```python
import cv2
import os
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
import numpy
from numpy import cov
from numpy import trace
from numpy import iscomplexobj
from numpy.random import random
from scipy.linalg import sqrtm
# Directory paths for generated fake images and real reference images
generated_images_dir = r'C:\Users\Manohar Vemuri\OneDrive\MS\Computer Vision\cs512-s23-vemuri-saimanohar\project\data\results\U-Net_CutMix_GAN/'
# real_images_dir = "D:\MS\sem2\cv\proj\00115.png"
real_image=cv2.imread(r"C:\images1024x1024\00001.jpg")
real_image = cv2.resize(real_image, (8, 8))
real_image = cv2.cvtColor(real_image, cv2.COLOR_BGR2GRAY)

# Iterate through all files in the generated images directory
a=[]
d=dict()

for file in os.listdir(generated_images_dir):
    if file.endswith('.png') or file.endswith('.jpg'):
        # Load the generated fake image
        fake_image = cv2.imread(os.path.join(generated_images_dir, file), cv2.IMREAD_GRAYSCALE)

        # Load the corresponding real reference image
        #print(real_image.shape)
        

        # Resize the images to a consistent size
        fake_image = cv2.resize(fake_image, (8, 8))
        act1=real_image
        act2=fake_image
      

        # Calculate SSIM
        mu1, sigma1 = act1.mean(axis=0), cov(act1, rowvar=False)
        mu2, sigma2 = act2.mean(axis=0), cov(act2, rowvar=False)
        # calculate sum squared difference between means
        ssdiff = numpy.sum((mu1 - mu2)**2.0)
        # calculate sqrt of product between cov
        covmean = sqrtm(sigma1.dot(sigma2))
        # check and correct imaginary numbers from sqrt
        if iscomplexobj(covmean):
            covmean = covmean.real
        # calculate score
        fid = ssdiff + trace(sigma1 + sigma2 - 2.0 * covmean)

        # Print the SSIM score for each pair of images
        a.append(fid)
        d[file]=fid
            
            

max_score=max(d.values())
min_score=min(d.values())
```

```python
s= list(d.keys())[list(d.values()).index(max_score)]
print(s)
sm=list(d.keys())[list(d.values()).index(min_score)]
print(sm)
```
